## PARTE I: Introducción a los repositorios

Los proyectos de programación tienen **miles de líneas de código** y **cientos de personas** trabajando en ellos, por lo que se necestian herramientas potentes para gestionar los cambios. Se usan los **repositorios de control de versiones**

Actualmente, dos de los más usados son: [Github](https://github.com/github) y [GitLab](https://gitlab.com/)
Ambos se basan en el sistema de control de versiones [GIT](https://en.wikipedia.org/wiki/Git), que es **software libre**, desarrollado en el 2005 por Linux Torvalds, el creador del kernel de Linux. Tanto Github como Gitlab tienen una **interfaz web** que hace más accesible su utilización

### Github

Github es uno de los repositorios más populares, que aloja más de **20 millones de repositorios de proyectos libres**. Se pueden encontrar estadísticas de github en [Octoverse](https://octoverse.github.com/)
